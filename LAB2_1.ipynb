### Write Python code to use Linear Regression to draw best fit line.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#basic example to see the linear regression term which is find the relation between 2 varaibles 
x =np.array([5,7,8,7,2,17,2,9,4,11,12,9,6]) # represnet Age 
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86]) # represent speed 
a, b = np.polyfit(x, y, 1)# finding the c and m slop and coff
plt.scatter(x, y)
plt.plot(x, a*x+b, color='steelblue', linestyle='--', linewidth=2)
plt.show()
```

### using predefined method from scipy

```python
#import stats to find slope and intercept 
from scipy import stats

x = [5,7,8,7,2,17,2,9,4,11,12,9,6] # represnt age 
y = [99,86,87,88,111,86,103,87,94,78,77,85,86] # represent speed 

slope, intercept, r, p, std_err = stats.linregress(x, y) #find slop and intercept 
#function to calculate the value of Y(predict value )
def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))
#plot the best fit line 
plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
```

```python
# another example 
#define data
x = np.array([1, 2, 3, 4, 5, 6, 7, 8])
y = np.array([2, 5, 6, 7, 9, 12, 16, 19])
```

```python
#find line of best fit
a, b = np.polyfit(x, y, 1)
```

```python
#add points to plot
plt.scatter(x, y, color='purple')
#add line of best fit to plot
plt.plot(x, a*x+b, color='steelblue', linestyle='--', linewidth=2)

#add fitted regression equation to plot
plt.text(1, 17, 'y = ' + '{:.2f}'.format(b) + ' + {:.2f}'.format(a) + 'x', size=14)
```
